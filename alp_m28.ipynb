Peça um número e informe se ele é maior que 10. Crie um algoritmo em Python para fazer essa verificação.

```python
# Solicita um número ao usuário e o converte para um valor inteiro
try:
    numero = int(input("Digite um número: "))

    # Verifica se o número é maior que 10
    if numero > 10:
        print(f"O número {numero} é maior que 10.")
    elif numero < 10:
        print(f"O número {numero} é menor que 10.")
    else:
        print(f"O número {numero} é igual a 10.")

except ValueError:
    print("Entrada inválida. Por favor, digite apenas números inteiros.")


```

Peça um número inteiro e informe se ele é par. Crie um algoritmo em Python que verifique se o número é divisível por 2.

```python
# Pede ao usuário para digitar um número inteiro
numero_str = input("Digite um número inteiro: ")

try:
    # Converte a entrada do usuário para um número inteiro
    numero = int(numero_str)

    # Verifica se o número é divisível por 2 (resto 0)
    if numero % 2 == 0:
        print(f"O número {numero} é par.")
    else:
        print(f"O número {numero} é ímpar.")

except ValueError:
    # Se a conversão falhar, informa que a entrada não é um número inteiro válido
    print("Entrada inválida. Por favor, digite um número inteiro.")
```

Peça um nome e informe se ele contém a letra "a". Crie um algoritmo em Python que faça essa verificação.

```python
# Solicita o nome ao usuário
nome = input("Por favor, digite um nome: ")

# Converte o nome para minúsculas para garantir que a verificação encontre 'a' e 'A'
nome_minusculo = nome.lower()

# Verifica se a letra 'a' está presente no nome
if 'a' in nome_minusculo:
    print(f"O nome '{nome}' contém a letra 'a'.")
else:
    print(f"O nome '{nome}' não contém a letra 'a'.")
```

Peça um ano e informe se ele é bissexto. Crie um algoritmo em Python que verifique se o ano é divisível por 4 e não por 100, exceto se também for divisível por 400.

```python
# Programa para verificar se um ano é bissexto

# Entrada: pedir o ano ao usuário
ano = int(input("Digite um ano: "))

# Processamento e saída
if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
    print(f"O ano {ano} é bissexto.")
else:
    print(f"O ano {ano} não é bissexto.")
```

Peça a quantidade de horas trabalhadas e o valor por hora. Se trabalhou até 40 horas, calcule o salário normalmente (horas × valor por hora). Se trabalhou mais de 40 horas, as horas extras devem ter um adicional de 50% (horas extras × valor por hora × 1.5). Crie um algoritmo em Python que realize esse cálculo corretamente.

```python
# Programa para calcular o salário com base nas horas trabalhadas

# Entrada: pedir os dados ao usuário
horas_trabalhadas = float(input("Digite a quantidade de horas trabalhadas: "))
valor_por_hora = float(input("Digite o valor pago por hora (R$): "))

# Processamento
if horas_trabalhadas <= 40:
    salario = horas_trabalhadas * valor_por_hora
else:
    horas_extras = horas_trabalhadas - 40
    salario = (40 * valor_por_hora) + (horas_extras * valor_por_hora * 1.5)

# Saída
print(f"Seu salário é: R${salario:.2f}")
```
